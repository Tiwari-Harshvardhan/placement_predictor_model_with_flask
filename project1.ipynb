```python
#we can store each test and train set into a different var to make it clear
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
#now we just need to scale the data, which is not so necessary here but it is a good proctice to do that 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
#now all the values are btw -1 to 1 but it is now in an array form so it is not easy to understand the row and column numbers as it was during splitting
#that is why we performed splitting first. but we need to always do splitting first to prevernt data leak
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
#notice that in the X_test we only used scaler.transform and not scaler.fit_transform as we understood the pattern while training itself
#now we will train the model
#we will use lgr(logical regression classifier
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
```

```python
#we have created an object called clf which can access the functions of logistic regression
```

```python
clf.fit(X_train,Y_train)
#this is model training
```

```python
#evaluate the model
```

```python
clf.predict(X_test)
```

```python
Y_test
```

```python
#we evaluate model using predict func of LogisticRegression
#we will pass it the new data or the test data and it will give us the output calculated by the model
```

```python
#we can calc the accuracy manually or using sklearn
```

```python
y_pred=clf.predict(X_test)
```

```python
Y_test
#Y_test and Y_pred have the same shape(linear tensor)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
#in metrics accuracy score we give the arguments as the original(True) data and the predicted data and it calcs the accuracy
```

```python
#how to plot decision boundary? that is what pattern the model recognized in our data we can plot that using a library called ml extend
#using this we can plot a graph we have to give the following arguments: independent training var(here X_train), dependent training var(Y_train), 
#classifier(clf in this case, LogisticRegression had we used it by its original name), and legend=2
```
